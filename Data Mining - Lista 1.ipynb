W zadaniu 1 należało źródło książki (w pliku .txt) "rozbić" na listę pojedynczych słów. W celu dokonania tego, każdą linijkę z książki muszę wczytać, rozdzielić na pojedyncze słowa i zapisać w liście ***words***: 

```python
from string import punctuation

words = []

stopwords = ['i', 'me', 'my',  'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 
'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 
'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 
'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 
'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 
'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 
'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now']

with open('OfMiceAndMen.txt', encoding='UTF-8') as f:
    words = [word for line in f for word in line.split()]
print(words)
    
```

Następnie należy usunąć znaki interpunkcyjne z otrzymanej listy słów i wszystkie litery w słowach zmienić na małe. Za te czynności odpowiada poniższy kod:

```python
words = [word.lower().translate(str.maketrans('','',punctuation)) for word in words]
print(words)
```

Potem trzeba usunąć *stopwords* (często występujące słowa, które nic nie wnoszą z punktu analizy tekstu źródłowego) i pozbyć się wszelkich napisów zawierających jakiekolwiek cyfry (nie traktuje ich jako słowa):

```python
filtered_words = [word for word in words if word not in stopwords]
filtered_words = [word for word in filtered_words if word.isalpha()]
print(filtered_words)
```

Następnym krokiem jest zastosowanie procedury stemmingu,w celu sprowadzenia każdego słowa do jego podstawowej formy ("słownikowej"). W tym celu można wykorzystać stemmery z dostępnych bibliotek, jednakże ich działanie bywa różne. W przypadku niektórych słów, stemmer może poprawnie sprowadzić rozpatrywane słowo do formy podstawowej, jednakże dla innych słów może np. ucinać końcówki przez co słowo nie jest "poprawnym" słowem. Jest to przypadłość większości ogólnie dostępnych stemmerów. Za tę czynność odpowiada następujący kod:  

```python
from stemming.porter2 import stem

words = [stem(word) for word in filtered_words]

print(words)
```

Po dokonaniu procedury stemmingu, tworzymy pary *(słowo, 1)*, które posłużą nam do zilczania wystąpień poszczególnych słów w tekście:

```python
pairs = [(word, 1) for word in words]
print(pairs)
```

Następnie będziemy zliczać pary, w której pierwszym elementem jest jakieś wyróżnione słowo. Liczba tych par jest 
dokładnie liczbą wystąpień tego słowa w tekście. W celu wykonania tego, niekoniecznie trzeba używać sortowania (bo dla dużych zbiorów danych może być kosztowne). Zamiast tego możemy użyć struktury danych zwanej słownikiem (dostępnej w Pythonie). Dzięki temu możemy zredukować złożoność tej procedury z $\mathcal{O}(n log(n))$ do $\mathcal{O}(n)$.

```python
# from itertools import groupby

# pairs.sort()

# word = lambda pair: pair[0]

# grouped_pairs = [(w, sum(1 for _ in g)) for w, g in groupby(pairs, key=word)]

occurences_dict = dict()

for p in pairs:
    if p[0] not in occurences_dict.keys():
        occurences_dict[p[0]] = 1
    else:
        occurences_dict[p[0]] += 1
        
    
grouped_pairs = list(occurences_dict.items())

print(grouped_pairs)
```

Następnie sortujemy pary malejąco względem liczby wystąpień w tekście:

```python
occurences = lambda pair: pair[1]

grouped_pairs.sort(key=occurences, reverse=True)

print(grouped_pairs)
```

Na koniec usuwam 10 najczęściej występujących słów i drukuje posortowaną listę w formie CSV (potrzebne do szybkiego utworzenia chmury punktów). Chmura punktów jest zgrupowaniem słów występujących w analizowanym tekście. Im większe i bardziej "pogrubione" jest słowo, tym częściej ono występuje w tekście (i jest ważniejsze).

```python
grouped_pairs = grouped_pairs[10:]

for p in grouped_pairs:
    print(f'{p[1]} ; {p[0]}')
```

W zadaniu 2 należało podzielić książkę na rozdziały i potraktować każdy rozdział jako osobny dokument. Pierwsze 2 podpunkty zadania 2 są dokładnie takie same jak w zadaniu 1 z tą różnicą, że listę słów dla każdego rozdziału trzymam w Pythonowym słowniku indeksowanym nazwami rozdziałów (dokumentów):

```python
document_words = dict()

documents = ['chapter1.txt', 'chapter2.txt', 'chapter3.txt', 'chapter4.txt', 'chapter5.txt', 'chapter6.txt']

for d in documents:
    
    with open(d, encoding='UTF-8') as f:
        words = [word for line in f for word in line.split()]
        
        words = [word.lower().translate(str.maketrans('','',punctuation)) for word in words]
        
        filtered_words = [word for word in words if word not in stopwords]
        filtered_words = [word for word in filtered_words if word.isalpha()]

        words = [stem(word) for word in filtered_words]
        
        document_words[d] = words
        
print(document_words['chapter1.txt'])
    
```

Następnie obliczam wagi $tf-idf$ dla każdego z dokumentów. Wzór na wagi wygląda następująco:
> $\mathcal tf-idf (t; d; D) = tf (t; d) \times idf (t; D)$

Gdzie $t$ jest rozpatrywanym słowem, $tf (t; d)$ jest liczbą wystąpień słowa $t$ w dokumencie $d$, $D$ jest kolekcją dokumentów do której należy dokument $d$ oraz $idf(t; D)$ to ***inverse document frequency***, który (krótko mówiąc) pokazuje nam jak dużo informacji niesie ze sobą dane słowo $t$. W poniższym kodzie znajduje się procedura obliczająca wagę $tf-idf$ dla danego słowa:

```python
from math import log10

def compute_tf_idf(term, document, collection):
    
    number_of_occurences = 0
    
    for word in collection[document]:
        if word == term:
            number_of_occurences += 1
            
    number_of_documents_with_occurence = 0
    
    for k, _ in collection.items():
        
        if term in collection[k]:
            number_of_documents_with_occurence += 1
            
    
    idf = log10(len(collection.keys()) / (number_of_documents_with_occurence))
    
    return idf * number_of_occurences

    
```

Następna procedura oblicza omawiane wagi dla każdego słowa w zadanym dokumencie:

```python
def compute_weights_for_document(document, collection):
    document_words = collection[document]
    used_words = set()
    
    pairs = []
    
    for word in document_words:
        if word not in used_words:
            
            tf_idf = compute_tf_idf(word, document, collection)
            
            pairs.append((word, tf_idf))
            
            used_words.add(word)
            
    comparator = lambda x: x[1]

    pairs.sort(key=comparator, reverse=True)
    
    return pairs

pairs = compute_weights_for_document('chapter6.txt', document_words)

print(pairs)
        
```

W ostatniej procedurze, wykorzystuje otrzymaną listę słów i odpowiadających im wag do wygenerowania chmury słów:

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


def generate_word_cloud(chapter_name, collection):
    
    pairs = compute_weights_for_document(chapter_name, collection)
    
    pairs = dict(pairs)
    
    word_cloud = WordCloud(width=500, height=400, mode='RGBA', background_color='white', max_words=200).fit_words(pairs)
    
    chapter_document_name = chapter_name.split('.')[0]
    
    plt.figure()
    plt.imshow(word_cloud, interpolation="bilinear")
    plt.axis("off")
    plt.savefig(f'{chapter_document_name}_wordcloud.png')
```

W zadaniu 3 należało zaimplementować funkcję, która jako argument przyjmuje słowo i zwraca posortowaną (malejąco) względem wag $tf-idf$ listę rozdziałów - dzięki takiemu rankingowi możemy stwierdzić, który rozdział najlepiej odpowiada danemu słowu (zasada działania trochę podobna do działania przeglądarki internetowej). Poniższa funkcja realizuje to zadanie:

```python
def generate_list_of_chapters(term, collection):
    chapters = list(collection.keys())
    
    rank_list = []
    
    for ch in chapters:
        weight = compute_tf_idf(term, ch, collection)
        
        rank_list.append((weight, ch))
        
    comparator = lambda x: x[0]
    
    rank_list.sort(key=comparator, reverse=True)
    
    return rank_list

chapter_ranking = generate_list_of_chapters('door', document_words)

print(chapter_ranking)
```

W zadaniu 4 należało dla każdego słowa utworzyć listę 5 najczęściej występujących po nim (bezpośrednio) słów (ale bez *stopwords*) a następnie wykorzystać ten ranking do wygenerowania *losowego* zdania, które przypomina zdanie z analizowanego tekstu. Poniższa funkcja dla każdego słowa w książce wyznacza liczbę wszystkich słów, które występują bezpośrednio po rozpatrywanym słowie:

```python

def five_most_common_words(words_list):
    statistic = dict()
    
    statistic = {word: dict() for word in words_list}
    
    for i in range(0, len(words_list)):
        current_word = words_list[i]
        
        next_word = f''
        
        if i + 1 < len(words_list):
            next_word = words_list[i + 1]
        
        if next_word in statistic[current_word].keys():
            statistic[current_word][next_word] += 1
        else:
            statistic[current_word][next_word] = 1
    
    for k, v in statistic.items():
        statistic[k] = list(v.items())
        
    for k, v in statistic.items():
        statistic[k] = statistic[k][:5]
    
    return statistic

ranking = five_most_common_words(filtered_words)
print(ranking)
```

Poniższa funkcja przyjmuje jako argument długość losowego zdania do wygenerowania oraz utworzony ranking najczęściej występujących po nim słów. Następnie funkcja wybiera losowo jedno ze słów występujących w rankingu (losowe słowo startowe). Potem wybiera losowo następne słowo (z prawdopodobieństwem proporcjonalnym do liczby wystąpień po rozważanym słowie) z listy 5 najczęściej występujących słów po słowie startowym i kontynuuje ten proces do momentu, w którym losowo generowane zdanie osiągnie zadaną długość: 

```python
from random import choice
from random import choices

def generate_random_paragraph(paragraph_length, words_ranking):
    words = list(words_ranking.keys())
    
    random_start_word = choice(words)
    
    random_paragraph = f'{random_start_word} '
    
    current_word = random_start_word
    
    for i in range(0, paragraph_length - 1):
        
        probability_weights = tuple([x[1] for x in words_ranking[current_word]])
        
        current_word = choices(words_ranking[current_word], probability_weights, k=1)[0][0]
       
        random_paragraph += f'{current_word} '
    
    return random_paragraph

print(generate_random_paragraph(10, ranking))
```

Uruchamiając powyższą funkcję widać, że części generowanych przez nią *losowych* zdań rzeczywiście znajdują się w oryginalnym tekście książki (aczkolwiek rzadko się zdarza, żeby pasowało np. całe zdanie). Wpływ na to ma np. to, że z tekstu usuneliśmy ***stopwords***, które są częścią poprawnych gramatycznie zdań oraz fakt, że patrzymy tylko na jedno słowo, występujące bezpośrednio po rozważanym słowie (a moglibyśmy patrzeć na $k$ najczęściej występujących słów po danym słowie oraz rozważyć to co występuje przed danym słowem). Krótko mówiąc, wybrany model jest trochę za słaby, aby wygenerować jakieś sensowne zdania.
